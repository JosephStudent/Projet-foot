## Analyse des statistiques des joueurs
En se basant sur les carte Stat FIFA.
je vais regrouper les données soudjascente entre elle et faire une moyenne ponderé.

PAC (Pace) : Moyenne de 'Acceleration' et 'Sprint Speed'
SHO (Shooting) : Moyenne de 'Finishing', 'Shot Power', 'Long Shots', 'Volleys', 'Penalties'
PAS (Passing) : Moyenne de 'Vision', 'Crossing', 'Free Kick Accuracy', 'Short Passing', 'Long Passing', 'Curve'
DRI (Dribbling) : Moyenne de 'Dribbling', 'Agility', 'Balance', 'Ball Control'
DEF (Defense) : Moyenne de 'Interceptions', 'Heading Accuracy', 'Def Awareness', 'Standing Tackle', 'Sliding Tackle'
PHY (Physical) : Moyenne de 'Jumping', 'Stamina', 'Strength', 'Aggression' 


```python
# Définir les catégories et les attributs correspondants
categories = {
    'PAC': ['Acceleration', 'Sprint Speed'],
    'SHO': ['Finishing', 'Shot Power', 'Long Shots', 'Volleys', 'Penalties'],
    'PAS': ['PAS','Vision', 'Crossing', 'Free Kick Accuracy', 'Short Passing', 'Long Passing', 'Curve'],
    'DRI': ['DRI','Dribbling', 'Agility', 'Balance', 'Ball Control'],
    'DEF': ['DEF','Interceptions', 'Heading Accuracy', 'Def Awareness', 'Standing Tackle', 'Sliding Tackle'],
    'PHY': ['PHY', 'Jumping', 'Stamina', 'Strength', 'Aggression']
}

# Fonction pour calculer les moyennes par catégorie
def calculate_means(df, categories):
    for category, attrs in categories.items():
        df[category] = df[attrs].mean(axis=1)
    return df

# Appliquer la fonction aux DataFrames
j_Masculin = calculate_means(j_Masculin, categories)
j_Feminin = calculate_means(j_Feminin, categories)

# Afficher les statistiques résumées pour les joueurs masculins et féminins
print("Données des joueurs Masculins :")
display(j_Masculin[['Name', 'Position', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY']].head())

print("Données des joueurs Féminins :")
display(j_Feminin[['Name', 'Position', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY']].head())
```

## Visualisation des performances par position
Cela nous permettra d'identifier les meilleurs joueurs pour chaque rôle.


```python
# Assure-toi que les colonnes utilisées pour l'agrégation sont numériques
cols_to_use = ['PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY', 'Positioning', 'Reactions']
j_masculin[cols_to_use] = j_masculin[cols_to_use].apply(pd.to_numeric, errors='coerce')

# Remplir les valeurs manquantes
j_masculin.fillna(j_masculin.mean(), inplace=True)

# Regrouper les données par position et calculer la moyenne
positions_data = j_masculin.groupby('Position')[cols_to_use].mean().reset_index()

# Créer les graphiques
plt.figure(figsize=(14, 10))
for i, category in enumerate(cols_to_use, start=1):
    plt.subplot(3, 3, i)
    sns.barplot(x='Position', y=category, data=positions_data)
    plt.title(f'Average {category} by Position')
    plt.ylabel('Average Score')
    plt.xlabel('Position')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Préparer les données par position
positions_data = j_Masculin.groupby('Position')[['PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY', 'Positioning','Reactions',  'Composure']].mean().reset_index()

# Créer les graphiques
plt.figure(figsize=(14, 10))
for i, category in enumerate(['PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY', 'Positioning','Reactions',  'Composure'], start=1):
    plt.subplot(3, 2, i)
    sns.barplot(x='Position', y=category, data=positions_data)
    plt.title(f'Average {category} by Position')
    plt.ylabel('Average Score')
    plt.xlabel('Position')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re  # Pour utiliser des expressions régulières

# Charger les données
data = pd.read_csv('female_players.csv')

# Fonction pour extraire les valeurs numériques de Height et Weight
def extract_numeric(value):
    if pd.isna(value):
        return None
    numeric_part = re.findall(r'\d+', value)
    return float(numeric_part[0]) if numeric_part else None

# Appliquer la fonction pour extraire et convertir les données
data['Height_cm'] = data['Height'].apply(extract_numeric)
data['Weight_kg'] = data['Weight'].apply(extract_numeric)

# Préparer les données par position
positions_data = data.groupby('Position')[['PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY', 'Positioning', 'Reactions', 'Composure', 'Height_cm', 'Weight_kg']].mean().reset_index()

# Créer les graphiques
plt.figure(figsize=(18, 15))  # Ajuste la taille pour accueillir les graphiques
attributes = ['PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY', 'Positioning', 'Reactions', 'Composure', 'Height_cm', 'Weight_kg']
for i, category in enumerate(attributes, start=1):
    plt.subplot(4, 3, i)
    sns.barplot(x='Position', y=category, data=positions_data)
    plt.title(f'Average {category} by Position')
    plt.ylabel('Average Score')
    plt.xlabel('Position')
    plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
